```python
import numpy as np
import matplotlib.pyplot as plt
np.set_printoptions(threshold=np.inf)
```

```python
interval = np.array(np.linspace(1,10, num = 1001))
```

```python
modInterval52 = interval
for i in range (0,52):
    modInterval52 = np.sqrt(modInterval52)
    
for i in range (0,30):
    modInterval52 = modInterval52**2
print(modInterval52)
```

```python
plt.plot(interval, interval, interval, modInterval52)
```
